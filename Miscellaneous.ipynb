```python
# Transpose a matrix without transpose method
def transpose(lst):
  if lst == []:
    print("Empty list. Can't transpose.")
  else:
    rows = len(lst)
    cols = len(lst[0])

    transposed = [[0 for _ in range(rows)] for _ in range(cols)]

    for i in range(rows):
      for j in range(cols):
        transposed[j][i] = lst[i][j]

  return transposed

transpose([[1,2,3],[4,5,6]])
```

```python
def whatToBuy(prices, quantities, restriction, names):

  if not prices or not quantities:
    return [[],0]

  avg_prices = sum(prices) / len(prices)
  avg_quant = sum(quantities) / len(quantities)

  item_price = 0

  eligible_items = []
  non_eligible_items = []

  for p, q, n in zip(prices, quantities, names):
    item_price = (p*q) / (avg_prices * avg_quant)

    if item_price < restriction:
      eligible_items.append(n)
    else:
      non_eligible_items.append(n)

  return [sorted(eligible_items), len(non_eligible_items)]

prices = [5, 2, 9, 3, 1, 2, 9]
quantities = [1, 1, 1, 1, 6, 6, 6]
restriction = 0.5
names = ['pencil', 'bottle', 'wallet', 'phone', 'toy', 'mouse', 'door']

result = whatToBuy(prices, quantities, restriction, names)
print(result)
```

```python
# Prime numbers
def isPrime(num):
    if num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                return False
                break
        else:
          return True
    else:
        return False

isPrime(11)
```

```python
# Minimum in array without min()
def min(arr):
  low = arr[0]
  for i in arr:
    if i < low:
      low = i
  return low

min([15,12,11,18,112])
```

```python
# Return value with maximum occurences in a list
def occurMax(a):
  num_dict = {}
  for i in a:
    num_dict[i] = a.count(i)

  return max(num_dict, key=num_dict.get)

occurMax([1,2,2,3,3,3,4])
```

```python
# Longest Common Prefix of a list of strings
def common_prefix(strs) -> str:
    prefix = ""
    for chars in zip(*strs):
        if len(set(chars)) > 1:
            break
        prefix += chars[0]
    return prefix

common_prefix(["maximum","maxin","maximus"])
```

```python
# List elements multiplication using recursion
def mult(lst):
  if len(lst) == 1:
    return lst[0]
  elif len(lst) == 0:
    return "Empty list"
  else:
    return lst[0]*mult(lst[1:])

mult([5,4,2,1])
```

```python
# Stack class in Python
class Stack:

    def __init__(self):
        self.stack = ()

    def push(self, element):
        self.stack += (element,)

    def top(self):
        return self.stack[-1]

    def pop(self):
        pop_item = self.stack[-1]
        self.stack = self.stack[:-1]
        return pop_item

    def size(self):
        return len(self.stack)

    def empty(self):
        if len(self.stack) == 0:
            return True
        else:
            return False

    def print_stack(self):
        print(self.stack)

    def reverse(self):
      reverse_stack = ()
      while self.size() > 0:
        popped_item = self.pop()
        reverse_stack += (popped_item,)
      return reverse_stack

    def min(self):
      minimum = min(self.stack)
      return minimum

    def max(self):
      maximum = max(self.stack)
      return maximum
```

```python
# Nearest smaller element to the left
def nse(a):
    st = Stack()
    result = []
    for num in a:
      while not st.empty():
        if st.top() >= num:
          st.pop()
        else:
          current_top = st.top()
          result.append(current_top)
          break
      if st.empty():
        result.append(-1)
      st.push(num)
    return result
```

```python
# Check for balanced parantheses using Stack()
def isBalanacedParentheses(s):
    open_tup = tuple('({[')
    close_tup = tuple(')}]')
    map = dict(zip(open_tup, close_tup))
    stack = Stack()

    for i in s:
        if i in open_tup:
            stack.push(map[i])
        elif i in close_tup:
            if stack.empty() or i != stack.pop():
                return False
    if stack.empty():
        return True
    else:
        return False

isBalanacedParentheses('{[]{()}}')
```

```python
isBalanacedParentheses('())')
```

```python
def isPalindrome(s):
    st = Stack()
    for i, letter in enumerate(s):
        st.push(letter)
        if st.top() == s[-i-1]:
            return True
        else:
            return False

isPalindrome('madam')
```

```python
# Queue class
class Queue:

    def __init__(self):
        self.queue = []

    def enqueue(self, inp):
        self.queue.append(inp)

    def dequeue(self):
        self.queue.pop(0)

    def front(self):
        return self.queue[0]

    def rear(self):
        return self.queue[-1]

    def size(self):
        return len(self.queue)

    def sum_queue(self):
        return sum(self.queue)

    def print_queue(self):
      print(self.queue)
```

```python
# Array reverse using queue
def reverse(a):
    que = Queue()
    reverse = []
    for num in a:
        que.enqueue(num)
    while que.size() > 0:
      reverse = [que.front()] + reverse
      que.dequeue()
    return reverse

reverse([1,2,3])
```

```python
# Max sum in sliding window with queue
def maxWindowSum(a, k):
  results = []
  queue = Queue()
  for i in a:
    queue.enqueue(i)
    if queue.size() == k:
      results.append(queue.sum_queue())
      queue.dequeue()
  return max(results)

maxWindowSum([2, 4, 6, 8, 10], 3)
```

```python
# Circular Queue class
class CircularQueue:

    def __init__(self, max_size):
        self.max_size = max_size
        self.queue = []

    def enqueue(self, inp):
        if self.size()+1 <= self.max_size:
          self.queue.append(inp)
        else:
          self.dequeue()
          self.queue.append(inp)

    def dequeue(self):
        self.queue.pop(0)

    def front(self):
        return self.queue[0]

    def rear(self):
        return self.queue[-1]

    def size(self):
        return len(self.queue)

    def sum_queue(self):
        return sum(self.queue)

    def print_queue(self):
      print(self.queue)
```

```python
# Hot potato game with queue
def hotPotato(namelist, num):
    q = CircularQueue(max_size = len(namelist))
    for name in namelist:
        q.enqueue(name)

    while q.size() > 1:
        for i in range(num):
            q.enqueue(q.front())
        q.print_queue()
        q.max_size -=1
        q.dequeue()

    return q.rear()

result = hotPotato(['John', 'Jane', 'Jim', 'Jill', 'Jeff', 'Jerry'], 5)
print(f"Hot potato winner is: {result}")
```

```python
# Adding a book to the catalog dictionary
def add_book(catalog, title, author, year):
    catalog[title] = {}
    catalog[title]['author'] = author
    catalog[title]['year'] = year

    if title in catalog:
        catalog[title]['available'] = True
    else:
        catalog[title]['available'] = False

add_book(catalog, 'The Lion, the Witch and the Wardrobe', 'C. S. Lewis', 1950)
```

```python
# Searching the catalog by title
def search_book(catalog, title):
    if title in catalog:
        return catalog[title]
    else:
        return 'Book not found in catalog.'

search_book(catalog, 'Brave New World')
```

```python
# Working with NumPy
# Inseting a value in an index of numpy array without .insert()
import numpy as np
def array_modifier(lst, value, index):
    temp = []
    ar = np.array(lst)
    for i in ar[len(ar):index-1:-1]:
        temp.append(i)
        ar = np.delete(ar, i-1)
    ar = np.append(ar, value)
    ar = np.append(ar, temp[::-1])
    return ar
```
